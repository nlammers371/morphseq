# Random 10x10 RdBu Grid


```python
from pathlib import Path

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

FIG_DIR = Path('/Users/nick/Library/CloudStorage/GoogleDrive-nlammers@uw.edu/My Drive/projects/K99/Materials/figure components/Aim 2')
FIG_DIR.mkdir(parents=True, exist_ok=True)

rng = np.random.default_rng(20260612)
grid = rng.normal(0, 0.45, size=(10, 10))

grid[1, 6] = 1.9
grid[2, 6] = 1.5
# grid[4, 5] = -2.0
# grid[5, 5] = -1.5

grid = np.flipud(grid)

mpl.rcParams.update({
    'figure.dpi': 150,
    'savefig.dpi': 300,
    'pdf.fonttype': 42,
    'ps.fonttype': 42,
    'savefig.bbox': 'tight',
    'savefig.facecolor': 'white',
})

fig, ax = plt.subplots(figsize=(4.0, 4.0))
mesh = ax.pcolormesh(
    np.arange(11),
    np.arange(11),
    grid,
    cmap='RdBu_r',
    vmin=-2,
    vmax=2,
    edgecolors='black',
    linewidth=1.0,
    shading='flat',
)

ax.set_aspect('equal')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('')
ax.set_ylabel('')
for spine in ax.spines.values():
    spine.set_visible(False)

cbar = fig.colorbar(mesh, ax=ax, fraction=0.046, pad=0.04)
cbar.set_ticks([])
cbar.set_label('')
cbar.outline.set_edgecolor('black')
cbar.outline.set_linewidth(1.0)

for suffix in ['png', 'pdf']:
    fig.savefig(FIG_DIR / f'random_10x10_rdBu_grid.{suffix}')

plt.show()

```
