```python
import pandas as pd
```

```python
mseq_path = "/net/trapnell/vol1/home/mdcolon/proj/morphseq/results/mcolon/20250310/data/ccs_mseq_pln_data_with_PCs.csv"

mseq_df = pd.read_csv(mseq_path)
```

```python
set(mseq_df.columns)
```

```python
mseq_df["dist_from_spline"]
mseq_df["PC1"]
```

```python
import matplotlib.pyplot as plt

# --- First figure: color by 'dist_from_spline' ---
plt.figure()
plt.scatter(mseq_df["PC1"], mseq_df["PC2"], c=mseq_df["dist_from_spline"])
plt.colorbar(label="dist_from_spline")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Colored by dist_from_spline")

# --- Second figure: color by 'perturbation' ---
# If 'perturbation' is categorical, we map each category to a different color
# automatically by plotting each category separately (without specifying colors).
plt.figure()
perturbations = mseq_df["perturbation"].unique()
for p in perturbations:
    subset = mseq_df[mseq_df["perturbation"] == p]
    plt.scatter(subset["PC1"], subset["PC2"], label=p)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Colored by perturbation")
plt.legend()

plt.show()


plt.figure()
experiment = mseq_df["expt"].unique()
for p in experiment:
    subset = mseq_df[mseq_df["expt"] == p]
    plt.scatter(subset["PC1"], subset["PC2"], label=p)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Colored by experiment")
plt.legend()

plt.show()
```
