### Extract latent space scores for each image

```python
mu_array = np.empty((len(data_loader), 50))
mu_array_norm = np.empty((len(data_loader), 50))
time_array = np.empty((len(data_loader), 1))
embryo_array = np.empty((len(data_loader), 1))
data_list = list(data_loader)

with torch.no_grad():
    for idx, data in enumerate(data_list):
        im_path = im_data.samples[idx][0]
        im_name = im_path.replace(datadir + 'class0/', '')
        im_name = im_name.replace('.tif', '')

        imgs, _ = data
        imgs = imgs.to(device)
        img = np.transpose(imgs[0].cpu().numpy(), [1, 2, 0])
        
        out, mu_array[idx, :], logVAR = net(imgs)
        
        time_array[idx] = int(im_name[-3:])
        
        embryo_array[idx] = int(im_name[-8:-5])
        
#         print(idx/len(data_list))
```

```python
# subtract first time point from each embryo
emb_list = np.unique(embryo_array)
for e in emb_list:
    emb_ids = np.where(embryo_array==e)[0]
    emb_mu = mu_array[emb_ids]
    emb_time = time_array[emb_ids]
    min_i = np.argmin(emb_time)
    mu_min = emb_mu[min_i, :]
    
    mu_array_norm[emb_ids, :] = mu_array[emb_ids, :].copy() - mu_min
```

```python
# subtract first time point from each embryo
embedding_norm2 = np.empty(embedding.shape)
emb_list = np.unique(embryo_array)
for e in emb_list:
    emb_ids = np.where(embryo_array==e)[0]
    emb_mu = embedding[emb_ids]
    emb_time = time_array[emb_ids]
    min_i = np.argmin(emb_time)
    mu_min = emb_mu[min_i, :]
    
    embedding_norm2[emb_ids, :] = embedding[emb_ids, :].copy() - mu_min
```
