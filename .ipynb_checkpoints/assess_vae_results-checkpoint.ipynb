```python
# save the model
import torch
import matplotlib.pyplot as plt
import numpy as np
import random
import os
from torchvision import datasets, transforms
from torchvision.utils import save_image
from torch.utils.data.sampler import SubsetRandomSampler

from model.conv_vae import VAE

```

### Load model 

```python
"""
Determine if any GPUs are available
"""
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# set path to model and data
model_name = '20230519_vae_z025'
zd = int(model_name[-3:])
datadir = "E:/Nick/Dropbox (Cole Trapnell's Lab)/Nick/morphSeq/data/vae_test/max_images_res014/"
modeldir = "E:/Nick/Dropbox (Cole Trapnell's Lab)/Nick/morphSeq/data/vae_test/" + model_name + "/"
outdir = modeldir + "max_images_res014_pd/" + model_name + "/"

if not os.path.isdir(outdir):
    os.makedirs(outdir)

batch_size = 1

# build data sampler
im_data = datasets.ImageFolder(datadir, transform=transforms.Compose([transforms.Grayscale(num_output_channels=1),
                                                                            transforms.ToTensor()]))
data_loader = torch.utils.data.DataLoader(im_data, batch_size=batch_size)

# Model class must be defined somewhere
net = VAE().to(device)
net.load_state_dict(torch.load(modeldir + model_name))
net.eval()
```

### Apply to a random sample image

```python
with torch.no_grad():
    for data in random.sample(list(data_loader), 1):

        imgs, _ = data
        imgs = imgs.to(device)
        img = np.transpose(imgs[0].cpu().numpy(), [1, 2, 0])
        

        plt.subplot(121)
        plt.imshow(np.squeeze(img))
        plt.xticks([])
        plt.yticks([])
        plt.title('Original')
        
        out, mu, logVAR = net(imgs)
        outimg = np.transpose(out[0].cpu().numpy(), [1, 2, 0])

        plt.subplot(122)
        plt.imshow(np.squeeze(outimg))

        plt.xticks([])
        plt.yticks([])
        plt.title('Predicted')
        plt.show()
```

### Extract latent space scores for each image

```python
mu_array = np.empty((len(data_loader), zd))
time_array = np.empty((len(data_loader), 1))
embryo_array = np.empty((len(data_loader), 1))
data_list = list(data_loader)

with torch.no_grad():
    for idx, data in enumerate(data_list):
        im_path = im_data.samples[idx][0]
        im_name = im_path.replace(datadir + 'class0/', '')
        im_name = im_name.replace('.tif', '')

        imgs, _ = data
        imgs = imgs.to(device)
        img = np.transpose(imgs[0].cpu().numpy(), [1, 2, 0])
        
        out, mu, logVAR = net(imgs)
        mu_array[idx, :] = mu.cpu()
        
        time_array[idx] = int(im_name[-3:])
        
        embryo_array[idx] = int(im_name[-8:-5])
        
#         print(idx/len(data_list))
```

### Use UMAP to visualize latent encodings of different images

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import umap 

reducer = umap.UMAP()
scaled_z = StandardScaler().fit_transform(mu_array)
embedding = reducer.fit_transform(scaled_z)
```

```python
import plotly.express as px

fig = px.scatter(x=embedding[:, 0], y=embedding[:, 1], color=time_array.flatten(), template='ggplot2', 
                 title='Latent space labeled by time')

fig.show()
```

```python
fig = px.scatter(x=embedding[:, 0], y=embedding[:, 1], color=embryo_array.flatten(), 
                      template='plotly_white', title='Latent space labeled by embryo')


fig.show()
```
